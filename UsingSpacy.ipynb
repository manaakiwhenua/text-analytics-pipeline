### Locations

```python
import difflib
```

```python
def find_locations(doc):
    places = []
    for x in doc.ents:
        if x.label_ is 'LOC':
            if doc[x.start].text == 'the' or doc[x.start].text == 'The':
                places.append([x.text[4:], x.text, x.start+1, x.end, x.sent])
            elif doc[x.start].text== 'Mt':
                places.append(['Mount' + x.text[2:], x.text, x.start+1, x.end, x.sent])
            elif doc[x.start].text== 'Mt.':
                places.append(['Mount' + x.text[3:], x.text, x.start+1, x.end, x.sent])              
            else:
                places.append([x.text, x. text, x.start, x.end, x.sent])
    
    df = pd.DataFrame(places,columns=['Location', 'Original Tokenised Text','Start', 'End','Sentence']
             )
#     Check possible locations in gazetteers for exact matches
    df['NZGazAnt']=[True if X in nzgazAntartica else False for X in df.Location ]
    df['NZGaz']=[True if (df['NZGazAnt'] is True or X in nzgaz) else False for X in df.Location ]
    df['ScarNZ']=[True if X in SCARNZnames else False for X in df.Location ]
    df['ScarGlobal']=[True if X in SCARGlobalnames else False for X in df.Location]
    df['GeoNamesNZ']=[True if X in GeoNamesNZ else False for X in df.Location]
    df['GeoNamesAnt']=[True if X in GeoNamesAnt else False for X in df.Location]
    df['GeoNames']=[True if (df['GeoNamesNZ'] is True or df['GeoNamesAnt'] is True or X in GeoNamesFiltered) else False for X in df.Location ]
    df['inNZ'] = df.NZGaz | df.NZGazAnt | df.GeoNamesNZ
    df['inAntarctica'] = df.ScarGlobal | df.NZGazAnt | df.GeoNamesAnt
    df['exactMatch'] = df.NZGaz | df.ScarGlobal | df.GeoNames
    
    
#     filter locations by those not found in Antarctica or New Zealand gazeteers 
    df2=df[(df['inAntarctica']==False) & (df['inNZ']==False)].copy()
    

#     look through these remaining locations (including those found only in GeoNames) for close matches 
#     eg. McMurdo Dry Valley v McMurdo Dry Valley or for partial matches eg. Ross Sea Region == Ross Sea 
   
    df2['Close_Match_NZGazAnt']=  [difflib.get_close_matches(X, nzgazAntartica, cutoff=0.9)[0]
                                                  if len(difflib.get_close_matches(X, nzgazAntartica, cutoff=0.9))>0
                                                  else np.nan 
                                                  for X in df2['Location']] 
    df2['Close_Match_NZGaz']=  [difflib.get_close_matches(X, nzgaz, cutoff=0.9)[0]
                                                      if len(difflib.get_close_matches(X, nzgaz, cutoff=0.9))>0
                                                      else np.nan 
                                                      for X in df2['Location']] 
    df2['Close_Match_ScarNZ']=  [difflib.get_close_matches(X, SCARNZnames, cutoff=0.9)[0] 
                                                      if len(difflib.get_close_matches(X, SCARNZnames, cutoff=0.9))>0
                                                      else np.nan 
                                                      for X in df2['Location']] 
    df2['Close_Match_ScarGlobal']=  [difflib.get_close_matches(X, SCARGlobalnames, cutoff=0.9)[0]
                                                      if (X !='Antarctic' and len(difflib.get_close_matches(X, SCARGlobalnames, cutoff=0.9))>0)
                                                      else np.nan 
                                                      for X in df2['Location']] 
    df2['Close_Match_GeoNamesNZ'] = [difflib.get_close_matches(X, GeoNamesNZ, cutoff=0.9)[0] 
                                                      if len(difflib.get_close_matches(X, GeoNamesNZ, cutoff=0.9))>0
                                                      else np.nan 
                                                      for X in df2['Location']] 
    df2['Close_Match_GeoNamesAnt']=  [difflib.get_close_matches(X, GeoNamesAnt, cutoff=0.9)[0]
                                                      if (X !='Antarctic' and len(difflib.get_close_matches(X, GeoNamesAnt, cutoff=0.9))>0)
                                                      else np.nan 
                                                      for X in df2['Location']]
    df2['Close_Match']= (df2.Close_Match_NZGazAnt.notna() | df2.Close_Match_NZGaz.notna() | 
                         df2.Close_Match_ScarNZ.notna() | df2.Close_Match_ScarGlobal.notna() | 
                         df2.Close_Match_GeoNamesAnt.notna())
    
    df2['PartialMatch_NZGazAnt']=getBiggestSubStringMatch(doc, df2, nzgazAntartica)
    df2['PartialMatch_NZGaz']=getBiggestSubStringMatch(doc, df2, nzgaz)
    df2['PartialMatch_ScarNZ']=getBiggestSubStringMatch(doc, df2, SCARNZnames)
    df2['PartialMatch_ScarGlobal']=getBiggestSubStringMatch(doc, df2, SCARGlobalnames)
    df2['PartialMatch_GeoNamesNZ']=getBiggestSubStringMatch(doc, df2, GeoNamesNZ)
    df2['PartialMatch_GeoNamesAnt']=getBiggestSubStringMatch(doc, df2, GeoNamesAnt)
    df2['PartialMatch'] = (df2['PartialMatch_NZGazAnt'].notna() | df2['PartialMatch_NZGaz'].notna() | 
                           df2['PartialMatch_ScarNZ'].notna() | df2['PartialMatch_ScarGlobal'].notna() | 
                           df2['PartialMatch_GeoNamesAnt'].notna())
#     df2['PartialMatch'] = [X if X==True else np.nan for X in df2['PartialMatch']]
    
#     merge the filtered dataframe (now with close and partial matches) with the unfiltered dataframe
    df_unified = pd.merge(df, df2, how='left')
    
#     redo the inNZ,inAntarctica and Found columns to include close and partial matches
    df_unified.drop(columns=['inNZ', 'inAntarctica'], inplace=True)
    df_unified['inNZ'] = (df_unified.NZGaz | df_unified.NZGazAnt | df_unified.GeoNamesNZ | 
                          df_unified.Close_Match_NZGaz | df_unified.Close_Match_NZGazAnt | 
                          df_unified.Close_Match_GeoNamesNZ | df_unified.PartialMatch_NZGaz | 
                          df_unified.PartialMatch_NZGazAnt | df_unified.PartialMatch_GeoNamesNZ)
    df_unified['inAntarctica'] = (df_unified.ScarGlobal | df_unified.NZGazAnt | df_unified.GeoNamesAnt |
                                  df_unified.Close_Match_ScarGlobal | df_unified.Close_Match_NZGazAnt | 
                                  df_unified.Close_Match_GeoNamesAnt | df_unified.PartialMatch_ScarGlobal | 
                                  df_unified.PartialMatch_NZGazAnt | df_unified.PartialMatch_GeoNamesAnt)
    df_unified['Found'] = df_unified.exactMatch | df_unified.Close_Match | df_unified.PartialMatch
    return df_unified
```

```python
def getBiggestSubStringMatch(document, df, gazetteer):
    matches=[]
    for row in df.index:
#         print('working on row', row)
        fullStart = df.loc[row,'Start']
        fullEnd = df.loc[row,'End']
        subLength = fullEnd-fullStart-1
#         print(fullEnd, fullStart, subLength)
        Found = False
        while subLength >1:
            subStart = fullStart
            subEnd = fullStart+subLength
            while subEnd<=fullEnd:
                subString = document[subStart:subEnd].text
#                 print('looking for', subString)
                if subString in gazetteer:
                    matches.append(subString)
                    Found=True,
#                     print('found it!')
                    break
                subStart+=1
                subEnd+=1
            if Found:
                    break
            else:
                subLength-=1
        if Found==False:
            matches.append(np.nan)
    return matches
```

### Finding Locations in Collection of Documents

Running through the folder of converted text documents we have so far to see what sort of extracted locations *are not* confirmed by either of the two gazetteers to see what refinement needs to be added. More gazetteers, and some refienment where a name begins with 'the' are most likely next steps.

```python
def findLocationsInAllDocs(folder, NLPmodel='large'):
    
        
    filelist = [x for x in os.listdir(folder) if x.endswith('txt')]
    df = pd.DataFrame()
    for filename in filelist:
        file = open(folder+'/'+ filename)
        doc = file.read()
        file.close()
        if NLPmodel == 'small':
            document = nlp(doc)
        elif NLPmodel == 'medium':
            document = nlp_md(doc)
        elif NLPmodel == 'large':
            document = nlp_lg(doc)
    
        df2 = find_locations(document)
        df2['Source'] = filename
        df = pd.concat([df,df2])
        df = df.reset_index(drop=True)
        
    return df
```

```python
from spacy.matcher import Matcher
```

```python
def findSpecies(doc):
    
    """
    Dictionary approach.
    
    This code runs through token by token looking to match Genus + species, 
    Genus + species + species, G. species etc based on words found in the genus 
    and species columns of the supplied list of species.
    
    Also finds examples where a known genus is followed by an unreginsied species 
    if that example also appears in an abbreviated form elsewhere. 
    """

    foundSpeciesList = []
    foundSpecies = ''
    foundSpeciestokens = []
    foundSpeciesDict ={}
    possiblesNames =[]
    possiblesTokens = []

    
    for token in doc:
        if str(token) in uniqueGenus:
            foundSpecies = foundSpecies + str(token)
            foundSpeciestokens.append(token)
            j=1
            onlyGenus = True
            while (doc[token.i+j].text in uniqueSpecies):
                foundSpecies += ' ' + str(doc[token.i+j])
                foundSpeciestokens.append(doc[token.i+j])
                j+=1
                onlyGenus = False
                
            key = ''
            
            if onlyGenus ==False:
                for tkn in foundSpeciestokens[:-1]:
                    key += tkn.text[0] + '. '
                key+= foundSpeciestokens[-1].text
                if key not in foundSpeciesDict:
                    foundSpeciesDict[key]=[foundSpecies, True]
#                     print('adding ', key, ': [', foundSpecies, True,'] to dict' )

                foundSpeciesList.append([foundSpecies, foundSpecies, True, token.i, token.i+j, token.idx, token.sent])
#                 print('DEBUG 1 - adding ',foundSpecies, foundSpecies, True, token.i, 'to DF')

                
            else:
                if doc[token.i+1].is_alpha:
                    candidate = token.text + ' ' + doc[token.i+1].text
                    if candidate not in possiblesNames:
                        possiblesNames.append(candidate)
                        possiblesTokens.append((token, doc[token.i+1]))

            foundSpecies=''
            foundSpeciestokens=[]


        elif token.shape_ == 'X.':
            foundSpeciestokens=[]
            foundSpecies = foundSpecies + str(token)
            foundSpeciestokens.append(token)
            j=1
            while ((doc[token.i+j].is_alpha and doc[token.i+j].text in uniqueSpecies) or doc[token.i+j].shape_ == 'x.'):
                foundSpecies += ' ' + str(doc[token.i+j])
                foundSpeciestokens.append(doc[token.i+j])
                j+=1
            if doc[token.i+j-1].text in uniqueSpecies:
#                 print(doc[token.i+j-1].text, 'in uniqueSpecies')
                if foundSpecies in foundSpeciesDict:
                    foundSpeciesList.append([foundSpecies, foundSpeciesDict.get(foundSpecies)[0], foundSpeciesDict.get(foundSpecies)[1], token.i, token.i+j, token.idx, token.sent])
#                     print('DEBUG 2 -adding ',foundSpecies, foundSpeciesDict.get(foundSpecies)[0], foundSpeciesDict.get(foundSpecies)[1],token.i, 'to DF')
                else:
#                     print("looking for", foundSpeciestokens)
                    matcher = Matcher(nlp.vocab)
                    pattern = [{"SHAPE": "Xxxxx", 
                                   "TEXT"[0]: foundSpeciestokens[0].text[0]}
                              ]
                    
                    for tkn in foundSpeciestokens[1:-1]:
                        pattern.append({"SHAPE": "xxxx", 
                                       "TEXT"[0]: tkn[0].text[0]})
                    pattern.append({"TEXT": foundSpeciestokens[-1].text})
#                     print("looking for", pattern)                
                    matcher.add("Species_Pattern", None, pattern)

                    matches = matcher(doc)
                    if len(matches)>0:
                        start = matches[0][1]
                        end =  matches[0][2]
                        matched_span = doc[start:end]
                        inList = False
                        if doc[start].text in uniqueGenus:
                            inList = True
#                             print(doc[start].text, 'in UniqueGenus')

                        foundSpeciesList.append([foundSpecies, matched_span.text, inList, token.i, token.i+j, token.idx, token.sent])
#                         print('DEBUG 3 -adding ',foundSpecies, matched_span.text, inList, token.i, 'to DF')

                        foundSpeciesDict[foundSpecies]=[matched_span.text, inList]

                        if inList==False:
                            for match_id, start, end in matches:
                                matched_span = doc[start:end]
                                foundSpeciesList.append([matched_span.text, matched_span.text, inList, start,end, doc[start].idx, doc[start].sent])
#                                 print('DEBUG 4 -adding ',matched_span.text, matched_span.text, inList, start, 'to DF')
            foundSpecies=''
            foundSpeciestokens=[]
                
#     print(possiblesTokens)
    for tkns in possiblesTokens:
        shortPattern = [{"SHAPE": "X.", "TEXT"[0]: tkns[0].text[0]},
                           {"TEXT": doc[tkns[0].i+1].text}
                          ]
        fullPattern = [{"SHAPE": "Xxxxx", "TEXT": tkns[0].text},
                           {"TEXT": doc[tkns[0].i+1].text}
                          ]
#         print('candidate = ', tkns)
#         print("looking for", shortPattern) 
        matcher = Matcher(nlp.vocab)
        matcher.add("abbreviated_Species_Pattern", None, shortPattern)
        matches = matcher(doc)
        if len(matches)>0:
#             print('found abbreviated version of candidate')
            matcher.add("full_Species_Pattern", None, fullPattern)
            
            matches = matcher(doc)
            for match_id, start, end in matches:
                matched_span = doc[start:end]
                foundSpeciesList.append([matched_span.text, tkns[0].text + ' '+ tkns[1].text, False, start, end, doc[start].idx, doc[start].sent])
#                 print('DEBUG 5 -adding ',matched_span.text, tkns[0].text + ' '+ tkns[1].text, False ,start, 'to DF')



    FoundSpeciesDF = pd.DataFrame(foundSpeciesList,columns=['Found as', 'Full name', 'InDefinitiveList', 'Start', 'End','PositionOfFirstToken','Sentence'])
    FoundSpeciesDF.sort_values('Start', inplace=True)
    FoundSpeciesDF.reset_index(inplace=True,drop=True)
    FoundSpeciesDF['InReferenceList']= [True if x in uniqueSpeciesStrings else False for x in FoundSpeciesDF['Full name']]
#     FoundSpeciesDF['InReferenceList']=FoundSpeciesDF['Full name']
    return FoundSpeciesDF
```

## Combining Species and Location into Matrix

```python
import re
```

```python
def extractSpeciesandLocations(tokenisedDoc):
    df_species = findSpecies(tokenisedDoc)
    df_locations = find_locations(tokenisedDoc)
    df_locations = df_locations[df_locations.Found==True] #only include locations that were confirmed
    df_speciesXlocation = pd.merge(df_species.assign(key=0), df_locations.assign(key=0), on='key', 
                               suffixes=('_species', '_location')).drop('key', axis=1)
    return df_speciesXlocation
```

```python
def extractSpeciesandLocationsfromFolderofDocuments(folder, NLPmodel='large'):
    filelist = [x for x in os.listdir(folder) if x.endswith('txt')]
    for filename in filelist:
        file = open(folder+'/'+ filename)
        doc = file.read()
        file.close()
        if NLPmodel == 'small':
            document = nlp(doc)
        elif NLPmodel == 'medium':
            document = nlp_md(doc)
        elif NLPmodel == 'large':
            document = nlp_lg(doc)
            
        df = extractSpeciesandLocations(document)
        newfilename = re.sub('.txt', '-matrix.csv', filename)
        df.to_csv(folder+'/'+ newfilename)
```

```python
def compareString (string, possibilities, threshold=0.8):
    options = []
    metrics = ['SequenceMatcher', 'Jaro', 'Gotoh', 'Smith-Waterman']
    for candidate in possibilities:
        scoreSM = SequenceMatcher(None, string, candidate).ratio()
        scoreJ = textdistance.jaro_winkler.normalized_similarity(string, candidate)
        scoreG = textdistance.gotoh.normalized_similarity(string, candidate) 
        scoreSW = textdistance.smith_waterman.normalized_similarity(string, candidate) 
        for i, score in enumerate([scoreSM,scoreJ,scoreG,scoreSW]):
            if score > threshold:
                options.append([candidate, score, metrics[i]])
    return options
```
